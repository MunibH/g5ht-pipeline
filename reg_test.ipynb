```python
import sys
import os
import glob
import tifffile
import numpy as np
from skimage.io import imread

import matplotlib.pyplot as plt
from tqdm import tqdm
import scipy.ndimage as ndi


from microfilm.microplot import microshow
from microfilm.microplot import Micropanel
from microfilm.microanim import Microanim
from microfilm.microanim import Microanimpanel
from microfilm.microplot import Microimage

FIXED_PTH = r'C:\Users\munib\POSTDOC\DATA\g5ht-free\20251028\date-20251028_time-1500_strain-ISg5HT_condition-starvedpatch_worm001\fixed.tif'
# MOVE_PTH = r'C:\Users\munib\POSTDOC\DATA\g5ht-free\20251028\date-20251028_time-1500_strain-ISg5HT_condition-starvedpatch_worm001\registered\1176.tif'
MOVE_PTH = r'C:\Users\munib\POSTDOC\DATA\g5ht-free\20251028\date-20251028_time-1500_strain-ISg5HT_condition-starvedpatch_worm001\registered_test\1176.tif'
WARP_PTH = r'C:\Users\munib\POSTDOC\DATA\g5ht-free\20251028\date-20251028_time-1500_strain-ISg5HT_condition-starvedpatch_worm001\warped\1176.tif'

fixed = ndi.zoom(tifffile.imread(FIXED_PTH)[:,1,:,:], zoom=(3,1,1)) # just the red channel (ZHW)
move = tifffile.imread(MOVE_PTH)[:,1,:,:] # just the red channel (ZHW)
warped = ndi.zoom(tifffile.imread(WARP_PTH)[:,1,:,:], zoom=(3,1,1)) # just the red channel (ZHW)

reg_stack = np.concatenate((fixed[np.newaxis,:,:,:],move[np.newaxis,:,:,:]), axis=0) # CZHW
reg_stack.shape

warp_stack = np.concatenate((fixed[np.newaxis,:,:,:],warped[np.newaxis,:,:,:]), axis=0)
```

```python
%matplotlib qt

mn = 0
mx = 600

anim1 = Microanim(
    data=fixed[np.newaxis,:,:,:], cmaps=['pure_magenta'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[mn,mx])
anim2 = Microanim(
    data=warped[np.newaxis,:,:,:], cmaps=['pure_cyan'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[mn,mx])
anim3 = Microanim(
    data=warp_stack, cmaps=['pure_magenta','pure_cyan'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[[mn,mx],[mn,mx]])


anim4 = Microanim(
    data=fixed[np.newaxis,:,:,:], cmaps=['pure_magenta'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[mn,mx])
anim5 = Microanim(
    data=move[np.newaxis,:,:,:], cmaps=['pure_cyan'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[mn,mx])
anim6 = Microanim(
    data=reg_stack, cmaps=['pure_magenta','pure_cyan'], flip_map=False, fig_scaling=5, dpi=300,
    rescale_type='limits', limits=[[mn,mx],[mn,mx]])

plt.close('all')

animpanel = Microanimpanel(rows=3, cols=2,)
animpanel.add_element(pos=[0,0], microanim=anim1)
animpanel.add_element(pos=[1,0], microanim=anim2)
animpanel.add_element(pos=[2,0], microanim=anim3)

animpanel.add_element(pos=[0,1], microanim=anim4)
animpanel.add_element(pos=[1,1], microanim=anim5)
animpanel.add_element(pos=[2,1], microanim=anim6)

animpanel.ui
```
